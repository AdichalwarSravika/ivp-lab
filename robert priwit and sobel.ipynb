```python
#robert
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Read the input image
input_image = cv2.imread('C:/Users/DS20/Desktop/image.png', cv2.IMREAD_GRAYSCALE)

# Display the input image
plt.figure(figsize=(6, 6))
plt.imshow(input_image, cmap='gray')
plt.title('Input Image')
plt.axis('off')
plt.show()

# Convert the image to double (float32)
input_image = np.float32(input_image)

# Pre-allocate the filtered_image matrix with zeros
filtered_image = np.zeros_like(input_image)

# Robert Operator Masks
Mx = np.array([[1, 0], [0, -1]])  # Gradient in X direction
My = np.array([[0, 1], [-1, 0]])  # Gradient in Y direction

# Edge Detection Process using Robert Operator
for i in range(input_image.shape[0] - 1):
    for j in range(input_image.shape[1] - 1):
        # Gradient approximations
        Gx = np.sum(Mx * input_image[i:i+2, j:j+2])
        Gy = np.sum(My * input_image[i:i+2, j:j+2])
        
        # Calculate magnitude of the gradient
        filtered_image[i, j] = np.sqrt(Gx**2 + Gy**2)

# Display the filtered image (gradient magnitude)
plt.figure(figsize=(6, 6))
plt.imshow(filtered_image, cmap='gray')
plt.title('Filtered Image (Gradient Magnitude)')
plt.axis('off')
plt.show()

# Define a threshold value
threshold_value = 100  # Adjust this threshold as needed
output_image = np.copy(filtered_image)

# Apply thresholding
output_image[output_image < threshold_value] = 0
output_image[output_image >= threshold_value] = 255

# Display the output image after thresholding (edge detection)
plt.figure(figsize=(6, 6))
plt.imshow(output_image, cmap='gray')
plt.title('Edge Detected Image')
plt.axis('off')
plt.show()
```

```python
#prewitt
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Read the input image in grayscale
input_image = cv2.imread('C:/Users/DS20/Desktop/image.png', cv2.IMREAD_GRAYSCALE)

# Display the input image
plt.figure(figsize=(6, 6))
plt.imshow(input_image, cmap='gray')
plt.title('Input Image')
plt.axis('off')
plt.show()

# Convert the image to float32
input_image = np.float32(input_image)

# Pre-allocate the filtered_image matrix with zeros
filtered_image = np.zeros_like(input_image)

# Prewitt Operator Masks
Mx = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])  # Gradient in X direction
My = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])   # Gradient in Y direction

# Edge Detection Process using Prewitt Operator
for i in range(1, input_image.shape[0] - 1):  # Start from 1 to avoid borders
    for j in range(1, input_image.shape[1] - 1):  # Start from 1 to avoid borders
        # Gradient approximations
        Gx = np.sum(Mx * input_image[i-1:i+2, j-1:j+2])  # Region around the pixel
        Gy = np.sum(My * input_image[i-1:i+2, j-1:j+2])
        
        # Calculate magnitude of the gradient
        filtered_image[i, j] = np.sqrt(Gx**2 + Gy**2)

# Display the filtered image (gradient magnitude)
plt.figure(figsize=(6, 6))
plt.imshow(filtered_image, cmap='gray')
plt.title('Filtered Image (Gradient Magnitude)')
plt.axis('off')
plt.show()

# Define a threshold value
threshold_value = 100  # Adjust this threshold as needed
output_image = np.copy(filtered_image)

# Apply thresholding
output_image[output_image < threshold_value] = 0
output_image[output_image >= threshold_value] = 255

# Display the output image after thresholding (edge detection)
plt.figure(figsize=(6, 6))
plt.imshow(output_image, cmap='gray')
plt.title('Edge Detected Image')
plt.axis('off')
plt.show()
```
