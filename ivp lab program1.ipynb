```python
#contrastrechung
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('C:/Users/DS20/Desktop/image.png')

# Define the min and max intensity values for the new range
min_intensity = 50
max_intensity = 200

# Apply contrast stretching
stretched_image = cv2.normalize(image, None, min_intensity, max_intensity, cv2.NORM_MINMAX)

# Show the result
plt.subplot(1,3,1)
plt.imshow(image)
plt.title('Original Image')
plt.subplot(1,3,2)
plt.imshow(stretched_image)
plt.title('Contrast Stretched')
stretch_image = cv2.normalize(image, None, min_intensity, max_intensity, cv2.NORM_INF)
plt.subplot(1,3,3)
plt.imshow(stretch_image)
plt.title('norm inf Image')
plt.show()
```

```python
# HISTOGRAM EQUALIZTION
import cv2
import matplotlib.pyplot as plt

# Load the image (color image)
image = cv2.imread('C:/Users/DS20/Desktop/image.png')

# Convert the image from BGR to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(gray_image)

# Show the original and equalized images
plt.subplot(121)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Grayscale Image')
plt.subplot(122)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.show()
```

```python
# Gamma Correctio(power law transformation)
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('C:/Users/DS20/Desktop/image.png')

# Apply Gamma Correction
gamma = 1.5  # Choose a gamma value (1.0 means no change)
gamma_corrected = np.array(255 * (image / 255) ** gamma, dtype='uint8')

# Show the result
plt.subplot(121)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.subplot(122)
plt.imshow(cv2.cvtColor(gamma_corrected, cv2.COLOR_BGR2RGB))
plt.title('Gamma Corrected')
plt.show()
```

```python
# Image sharpening
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('C:/Users/DS20/Desktop/image.png')

# Define sharpening kernel
sharpening_kernel = np.array([[0, -1, 0],
                              [-1, 5,-1],
                              [0, -1, 0]])

# Apply filter
sharpened_image = cv2.filter2D(image, -1, sharpening_kernel)

# Show the result
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(122), plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB)), plt.title('Sharpened Image')
plt.show()
```

```python
#EDGE DETECTION
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('C:/Users/DS20/Desktop/image.png')

# Apply Canny edge detection
edges = cv2.Canny(image, threshold1=100, threshold2=200)

# Show the result
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(edges, cmap='gray'), plt.title('Edge Detection')
plt.show()
```

```python
#SMOOTHNING
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('C:/Users/DS20/Desktop/image.png')

# Apply Gaussian Blur
blurred_image = cv2.GaussianBlur(image, (5, 7), 2)

# Show the result
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(122), plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)), plt.title('Blurred Image')
plt.show()
```

```python
#NOISE REDUCTION
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('C:/Users/DS20/Desktop/image.png')

# Apply Non-Local Means Denoising
denoised_image = cv2.fastNlMeansDenoisingColored(image, None, 10, 10, 7, 21)

# Show the result
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(122), plt.imshow(cv2.cvtColor(denoised_image, cv2.COLOR_BGR2RGB)), plt.title('Denoised Image')
plt.show()
```

```python
#COLOR ENHANCEMENT
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('C:/Users/DS20/Desktop/image.png')

# Convert to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Increase the saturation (2x)
hsv_image[..., 1] = hsv_image[..., 1] * 2
hsv_image[..., 1] = np.clip(hsv_image[..., 1], 0, 255)  # Ensure values remain in valid range

# Convert back to BGR
enhanced_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)

# Show the result
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(122), plt.imshow(cv2.cvtColor(enhanced_image, cv2.COLOR_BGR2RGB)), plt.title('Color Enhanced')
plt.show()
```
